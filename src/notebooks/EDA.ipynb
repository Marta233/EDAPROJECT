```python
import pandas as pd

def load_data(file_path):
    """Load data from a CSV file."""
    return pd.read_csv('../../dataset/benin-malanville.csv')

def basic_info(data):
    """Print basic information about the DataFrame."""
    print("Basic Information:")
    display(data.info())

def calculate_summary_stats(data):
    """Calculate summary statistics for each numeric column."""
    print("Summary Statistics:")
    
    # Calculate summary statistics
    summary_stats = data.describe()
    
    # Calculate median
    median = data.median(numeric_only=True)
    median.name = 'median'
    
    # Calculate standard deviation
    std = data.std(numeric_only=True)
    std.name = 'std'
    
    # Calculate skewness
    skew = data.skew(numeric_only=True)
    skew.name = 'skew'
    
    # Calculate kurtosis
    kurtosis = data.kurt(numeric_only=True)
    kurtosis.name = 'kurtosis'
    
    # Concatenate summary statistics
    summary_stats = pd.concat([summary_stats, median, std, skew, kurtosis], axis=0)
    
    # Format the output
    summary_stats = summary_stats.rename({
        "50%": "median",
        "std": "standard deviation",
        "skew": "skewness",
        "kurtosis": "kurtosis"
    })
    
    return summary_stats
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
# 2. Outliers
plt.figure(figsize=(16, 12))

# Define a function to plot boxplots for each numeric attribute
def plot_boxplots(attribute, subplot_position):
    plt.subplot(4, 5, subplot_position)
    sns.boxplot(y=data[attribute])
    plt.title(attribute)

# Plot boxplots for each numeric attribute
numeric_attributes = data.select_dtypes(include=np.number).columns
for i, attribute in enumerate(numeric_attributes, start=1):
    plot_boxplots(attribute, i)
plt.tight_layout()
plt.show()
```

```python
"""Plot time series for specified columns."""
# Convert 'Timestamp' column to datetime format
data['Timestamp'] = pd.to_datetime(data['Timestamp'])
    
# Set 'Timestamp' column as index
data.set_index('Timestamp', inplace=True)
    
# Define colors
colors = sns.color_palette("husl", len(data.columns))
    
# Plot time series
plt.figure(figsize=(12, 6))
for i, col in enumerate(['GHI', 'DNI', 'DHI', 'Tamb']):  # Specify columns to plot
    sns.lineplot(data=data, x=data.index, y=col, label=col, color=colors[i], linewidth=2, alpha=0.8)
    plt.xlabel('Time', fontsize=12)
    plt.ylabel('Value', fontsize=12)
    plt.title('Time Series Analysis', fontsize=16)
    plt.legend(fontsize=10)
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    plt.tight_layout()
    plt.show()
```

```python
"""Perform correlation analysis between solar radiation and temperature variables."""
# Select relevant columns for correlation analysis
relevant_columns = ['GHI', 'DNI', 'DHI', 'TModA', 'TModB']
relevant_data = data[relevant_columns]
    
# Calculate correlation matrix
correlation_matrix = relevant_data.corr()
# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Wind Analysis: Explore wind speed (WS, WSgust, WSstdev) and wind direction (WD, WDstdev) data to identify any trends or notable wind events.
"""Explore wind speed and wind direction data."""
# Basic statistics
print("Basic Statistics:")
print(data[['WS', 'WSgust', 'WSstdev', 'WD', 'WDstdev']].describe())
    
# Visualize wind speed over time
plt.figure(figsize=(12, 6))
sns.lineplot(data=data, x='Timestamp', y='WS', label='Wind Speed (m/s)')
sns.lineplot(data=data, x='Timestamp', y='WSgust', label='Wind Gust Speed (m/s)')
sns.lineplot(data=data, x='Timestamp', y='WSstdev', label='Wind Speed Std Dev (m/s)')
plt.xlabel('Timestamp')
plt.ylabel('Wind Speed (m/s)')
plt.title('Wind Speed Analysis')
plt.legend()
plt.show()
```

```python
# Temperature Analysis: Compare module temperatures (TModA, TModB) with ambient temperature (Tamb) to see how they are related or vary under different conditions.

# """Compare module temperatures with ambient temperature."""
# Basic statistics
print("Basic Statistics:")
print(data[['TModA', 'TModB', 'Tamb']].describe())
    
# Visualize temperature trends
plt.figure(figsize=(12, 6))
sns.lineplot(data=data, x='Timestamp', y='TModA', label='Module A Temperature (°C)')
sns.lineplot(data=data, x='Timestamp', y='TModB', label='Module B Temperature (°C)')
sns.lineplot(data=data, x='Timestamp', y='Tamb', label='Ambient Temperature (°C)')
plt.xlabel('Timestamp')
plt.ylabel('Temperature (°C)')
plt.title('Temperature Analysis')
plt.legend()
plt.show()
    
# Calculate correlation
correlation = data[['TModA', 'TModB', 'Tamb']].corr()
print("\nCorrelation Matrix:")
print(correlation)
```
