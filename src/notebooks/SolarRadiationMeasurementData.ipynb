```python
import pandas as pd

def load_data(file_path):
    """Load data from a CSV file."""
    return pd.read_csv('../../dataset/benin-malanville.csv')

def basic_info(data):
    """Print basic information about the DataFrame."""
    print("Basic Information:")
    display(data.info())

def calculate_summary_stats(data):
    """Calculate summary statistics for each numeric column."""
    print("Summary Statistics:")
    
    # Calculate summary statistics
    summary_stats = data.describe()
    
    # Calculate median
    median = data.median(numeric_only=True)
    median.name = 'median'
    
    # Calculate standard deviation
    std = data.std(numeric_only=True)
    std.name = 'std'
    
    # Calculate skewness
    skew = data.skew(numeric_only=True)
    skew.name = 'skew'
    
    # Calculate kurtosis
    kurtosis = data.kurt(numeric_only=True)
    kurtosis.name = 'kurtosis'
    
    # Concatenate summary statistics
    summary_stats = pd.concat([summary_stats, median, std, skew, kurtosis], axis=0)
    
    # Format the output
    summary_stats = summary_stats.rename({
        "50%": "median",
        "std": "standard deviation",
        "skew": "skewness",
        "kurtosis": "kurtosis"
    })
    
    return summary_stats
```
